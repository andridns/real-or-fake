```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import string
DATASET = 'bulldog_e'
```

```python
cmt_thr = 30
auth_thr = 8
pred = True
f = open(f'{DATASET}_pred.txt',encoding='utf-8').read()

lines = [l[11:] for l in f.split('\n') if 'Comment by' in l 
         and len(l[l.find(':')+2:]) > cmt_thr
         and len(l[:l.find(':')]) > auth_thr]
authors = [l[:l.find(':')] for l in lines]
comments = [l[l.find(':')+2:] for l in lines]
dfp = pd.DataFrame(np.array([authors,comments,lines]).T, columns=['author','comment','line'])
dfp['true'] = 0 if pred else 1
dfp.shape
```

```python
dfp.drop_duplicates(inplace=True)
dfp.shape
```

```python
cmt_thr = 30
auth_thr = 8
pred = False
f = open(f'{DATASET}.txt',encoding='utf-8').read()
print('string length before :',len(f))
f = ''.join([st for st in f if st in string.printable])
print('string length after :',len(f))

lines = [l[11:] for l in f.split('\n') if 'Comment by' in l 
         and len(l[l.find(':')+2:]) > cmt_thr
         and len(l[:l.find(':')]) > auth_thr]
authors = [l[:l.find(':')] for l in lines]
comments = [l[l.find(':')+2:] for l in lines]
dfy = pd.DataFrame(np.array([authors,comments,lines]).T, columns=['author','comment','line'])
dfy['true'] = 0 if pred else 1
dfy.shape
```

```python
dfy.drop_duplicates(inplace=True)
dfy.shape
```

```python
# shuffle true label and randomly sample 1:1 class composition
dfy_shuf = dfy.sample(n=dfp.shape[0], random_state=42)
df = dfp.append(dfy_shuf, ignore_index=True)
# shuffle combined df rows
df = df.sample(frac=1).reset_index(drop=True)
df.shape, dfp.shape
```

```python
df.to_csv(f'{DATASET}.csv')
```
